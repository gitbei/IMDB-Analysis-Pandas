#### Most productive directors

```python
data.head()
```

```python
data_agg = data.groupby("director_name")[["year","title"]].agg({"year":["min","max"],"title":"count"})
```

```python
data_agg.head()
```

```python
data_agg.columns
```

```python
changed_columns = []
for i in data_agg.columns:
    #print(i)
    changed_columns.append("_".join(i))
print(changed_columns)
```

```python
data_agg.columns = changed_columns
```

```python
data_agg.head()
```

```python
data_agg.reset_index(inplace=True)
```

```python
data_agg.head()
```

```python
data_agg["active_years"]  = data_agg["year_max"] - data_agg["year_min"]
```

```python
data_agg.head()
```

```python
data_agg["productivity"] = (data_agg["title_count"]/data_agg["active_years"]).round(2)
```

```python
data_agg.head()
```

```python
data.drop("productivity",inplace=True,axis=1)
```

```python
data.head()
```
